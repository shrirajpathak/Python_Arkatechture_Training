```python
import xml.etree.ElementTree as ET
import xml.etree.ElementTree as ETree

import pandas as pd
```

```python
pd.options.display.max_rows = 200
```

```python
# 1. Read the above xml file using xml module specifically xml.etree.ElementTree 
df = pd.read_xml("uni_info.xml") 
df
```

```python
#create a df using pandas which will have the columns as follows "Stream", "Name", "Age", "Gender", "Roll_No" 
root = ET.parse("uni_info.xml")

data = []
for stream in root.findall('stream'):
    course = stream.find('course').text
    students = stream.findall('student')
    for student in students:
        name = student.find('name').text
        age = int(student.find('age').text)
        gender = student.find('gender').text
        roll_number = student.find('roll_number').text
        data.append([course, name, age, gender, roll_number])

df = pd.DataFrame(data, columns=['stream', 'name', 'age', 'gender', 'roll_number'])
```
